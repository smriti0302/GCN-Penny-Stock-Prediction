```python
#close, high, low, open, volume
stock_price_data = np.load(r"C:\Users\Sivakumar\Desktop\GCN_stock_price_prediction-main\Sample_Dataset_code\stock_price_matrix_adjust.npy")
stock_price_data
```

```python
stock_price_data.shape
```

```python
ad_sample_data = stock_price_data
ad_sample_data[np.isnan(ad_sample_data)] = 5
```

```python
ad_sample_data
```

```python
Pearson_stock_price = np.zeros((ad_sample_data.shape[1],ad_sample_data.shape[1]))
Pearson_stock_price.shape
```

```python
#(N,N) - Pearson
for i in range(ad_sample_data.shape[1]):
    x = ad_sample_data[:,i,0]
    for j in range(ad_sample_data.shape[1]):
        y = ad_sample_data[:,j,0]
        pccs = np.corrcoef(x, y)
        Pearson_stock_price[i,j] = pccs[0,1]
Pearson_stock_price
```

```python
Pearson_stock_price[np.isnan(Pearson_stock_price)] = 0
```
